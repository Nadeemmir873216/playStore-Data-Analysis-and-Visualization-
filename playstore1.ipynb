## Visulaizing data

```python
df = pd.read_csv("cleaned_data.csv")
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')
df.dropna(subset=['Reviews'], inplace=True)
```

### Top 5 apps with most reviews

```python
top_apps = df.nlargest(10, 'Reviews')
```

```python

# Create a bar plot
plt.figure(figsize=(12, 6))
plt.barh(top_apps['App'], top_apps['Reviews'])


plt.title('Top 10 Apps with Most Reviews on Play Store')
plt.xlabel('Number of Reviews')
plt.ylabel('App Name')

# Show the plot
plt.gca().invert_yaxis()
```

### Top 5 apps with best ratings

```python
top_apps_ratings = df.nlargest(5, 'Rating')

# Create a horizontal bar plot
plt.figure(figsize=(12, 6))
plt.barh(top_apps_ratings['App'], top_apps_ratings['Rating'], color='lightgreen')

# Add titles and labels
plt.title('Top 10 Apps with Best Ratings on Play Store')
plt.xlabel('Rating')
plt.ylabel('App Name')

# Invert y-axis to have the app with the highest rating on top
plt.gca().invert_yaxis()
```

### Distribution of rating on Store

```python
plt.hist(df['Rating'], bins=20, color='purple', alpha=0.7)
plt.title('Distribution of App Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

### Category counts

```python
category_counts = df['Category'].value_counts()
plt.figure(figsize=(12, 6))
plt.bar(category_counts.index, category_counts.values, color='orange')
plt.xticks(rotation=90)
plt.title('Number of Apps by Category')
plt.xlabel('Category')
plt.ylabel('Number of Apps')
plt.show()
```

### Reviews vs ratings

```python
plt.scatter(df['Reviews'], df['Rating'], alpha=0.1, color='teal')
plt.title('Reviews vs. Ratings')
plt.xlabel('Number of Reviews')
plt.ylabel('Rating')
plt.xscale('log')  # Optional: Use logarithmic scale for better visibility
plt.show()
```

### Pie chart for 5 categories

```python
top_categories = df['Category'].value_counts().head(5)
plt.pie(top_categories, labels=top_categories.index, autopct='%1.1f%%', startangle=90)
plt.title('Top 5 App Categories by Count')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular
plt.show()
```
